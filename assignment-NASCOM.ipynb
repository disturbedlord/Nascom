# e. Consider the correlation between Spreadsheet and Proofreading.

i.  What is the value of the correlation coefficient r for this correlation?

```r
spreadsheet <- file$Spreadsheet
proofreading <- file$Proofreading

coeff <- cor(spreadsheet , proofreading)
coeff
```

ii.  What is the value of tau?

```r
tau <- cor.test(spreadsheet , proofreading , method = "kendall" , exact = FALSE)
tau$estimate
```

iii.  What is the value of rho?

```r
rho <- cor.test(spreadsheet , proofreading , method = "spearman" , exact = FALSE)
rho$estimate
```

# f.  Conduct a linear regression of Proofreading vs. Words.per.minute.

```r
wordsPerMinute = file$Words_per_minute

linearModal <- lm(wordsPerMinute ~ proofreading)
summary(linearModal)
```

i.  What is the p-value for this model?

```r
pValue <- anova(linearModal)$'Pr(>F)'[1]
pValue
```

ii.  What is the r-squared value?

```r
rSquared <- summary(linearModal)$r.squared
rSquared
```

```r
scatter.smooth(wordsPerMinute , proofreading , main = "Words_per_minute V/S Proofreading")
```

iii.  Do the residuals suggest that the linear regression model is an appropriate model?

```r
plot(linearModal)
```
